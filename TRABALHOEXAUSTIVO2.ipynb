```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("catherinerasgaitis/mxmh-survey-results")
```

```python
import os
import pandas as pd # Import the pandas library and assign it the alias 'pd'

df = pd.read_csv(os.path.join(path, "mxmh_survey_results.csv"))
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset=["Fav genre", "Hours per day", "Anxiety", "Depression"])
df = df[(df["Hours per day"] >= 0) & (df["Hours per day"] <= 24)]
df = df[(df["Anxiety"] >= 0) & (df["Anxiety"] <= 10)]
df = df[(df["Depression"] >= 0) & (df["Depression"] <= 10)]
```

```python
df.to_csv("dadosdash.csv", index=False)
```

```python
print(df.dtypes)
print(df[['Hours per day', 'Anxiety', 'Depression']].describe())
print(df[df['Hours per day'] > 24])
```
